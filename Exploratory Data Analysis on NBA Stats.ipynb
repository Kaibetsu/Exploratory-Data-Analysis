## Python Exploratory Data Analysis

### Web scraping using pandas

Import library

```python
import pandas as pd
```

The "2018-19 NBA Player Stats: Per Game" data will be retrieved from http://www.basketball-reference.com/.

Retrieve HTML table data

```python
html = pd.read_html('https://www.basketball-reference.com/leagues/NBA_2019_per_game.html', header = 0)
df2019 = html[0]
```

Clean the data

```python
raw = df2019.drop(df2019[df2019.Age == 'Age'].index)
raw
```

### Data Cleaning

#### Data Dimensions

```python
raw.shape
```

#### Dataframe Contents

```python
raw.head()
```

```python
#check for missing values
raw.isnull().sum()
```

```python
#replace missing values with 0
df = raw.fillna(0)
df.isnull().sum()
```

```python
#drop extra index column
df = df.drop(['Rk'], axis=1)
```

```python
#preview new df
df
```

#### Write to csv file

```python
df.to_csv('nba2019.csv', index=False)
```

## Posing Questions

#### Conditional Selection

What are the stats of the player that scored the most points? (PTS)

```python
df[df.PTS == df.PTS.max()]
```

We can further analyze this player utilizing conditional selection.
The next question will be: What team is this player from? (Tm)

```python
MostPointsPlayer = df[df.PTS == df.PTS.max()]
MostPointsPlayer.Tm
```

What position do they play? (Pos)

```python
MostPointsPlayer.Pos
```

How many games did they play during the season? (G)

```python
MostPointsPlayer.G
```

#### Show the stats for every player that scored more than 20 points per game.

```python
df[df.PTS > 20]
```

#### Which player scored the most 3 pointers per game? (3P)

```python
df[df['3P'] == df['3P'].max()]
```

#### Which player had the most assists per game? (AST)

```python
df[df['AST'] == df['AST'].max()]
```

### Using the groupby() function

#### Which player scored the most points on the Los Angeles Lakers?

```python
#create variable LAL for the team
LAL = df.groupby('Tm').get_group('LAL')
```

```python
LAL[LAL.PTS == LAL.PTS.max()]
```

#### To expand on the highest scoring player question, is it common for players in his position (point guard) to be scoring the most points? Which position typically is the highest scoring?

```python
#first we must group players by position
df.groupby('Pos').PTS.describe()
```

```python
#let's analyze only the players that have a single position
positions = ['C','PF','SF','PG','SG']
POS = df[df['Pos'].isin(positions)]
POS
```

```python
#now let's look at the points scored by position!
POS.groupby('Pos').PTS.describe()
```

This reveals that the highest scoring position is the center! The point guard is a very close second.

### Heat Map

#### Compute the correlation matrix

```python
corr = df.corr()
corr
```

#### Visualize using heatmaps

```python
sns.heatmap(corr)
```

```python
fig, ax = plt.subplots(figsize=(7,5))
sns.heatmap(corr, square=True)
```

This looks very messy. We can mask the top diagonal half of the heatmap to provide it with a slightly cleaner visual.

```python
# https://seaborn.pydata.org/generated/seaborn.heatmap.html
import numpy as np

mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(7, 5))
    ax = sns.heatmap(corr, mask=mask, vmax=1, square=True)
```
